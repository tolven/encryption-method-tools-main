```python
def elGamal(p,g,a,b,M):
    print("Alice a: ", a)
    print("A = g^a mod p")
    print("A = ",g,"^",a," mod ", p)
    A = (g**a) % p
    print("A = ", A)
    print("Nachricht an B(öffentlicher Schlüssel): ")
    print("(A=", A, ",p=",p, "g=",g,")")
    print("\n")
    
    print("Bob b: ", b)
    print("Verschlüsseln der Nachricht M = ", M)
    print("B = g^b mod p")
    print("B = ",g,"^",b," mod ", p)
    B = (g**b) % p
    print("B = ", B)
    print("\n")
    print("C = A^b * M mod p")
    print("C = ", A, "^", b, "*", M, " mod ", p)
    C = A**b * M % p
    print("C = ", C)
    print("Verschlüsselte Nachricht an A: ")
    print("(B = ", B, "C = ", C,")")
    print("\n")
    print("Entschlüsseln bei Alice a: ", a)
    print("M = C * B^-a mod p = C * B^p-1-a mod p")
    print("M = ", C, "*", B,"^(",p,"-",1,"-",a,") mod ", p)
    print("M = ", C, "*", B,"^(",p-1-a,") mod ", p)
    M = (C * (B ** (p-1-a))) % p
    print("M = ", M)
```

```python
elGamal(23,7,6,13,17)
```
