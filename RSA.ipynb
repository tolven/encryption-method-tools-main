```python
import math
import random as rnd
```

```python
def euler(p,q):
    e_n = ((p-1) * (q-1))
    return e_n
```

```python
def find_e(e_n):
    e = 1
    y = 0
    e_s = []
    while(len(e_s) <= 50):
        e += 1
        y = math.gcd(e,e_n)
        if(y == 1):
            e_s.append(e)
    random_int = rnd.randint(0,len(e_s)-1)
    print(y , " = ", e_s[random_int], " mod ", e_n)
    print("e = ggt(e,φ(n) = 1 -> ggt(", e_s[random_int],",",e_n,") = 1")
    return e_s[random_int]
```

```python
def find_d(e, e_n):
    d = 1
    temp = 0
    
    while(temp != 1):
        d += 1
        temp = e * d % e_n
        print(temp, " = ", e, " * ", d, " mod ", e_n)
    print("d = e * d = 1 mod φ(n) -> ",e, "*", d, " = 1 * ",e_n)
    return d
```

```python
def crypt(msg, e, d,n):
    ver = (msg**e) % n
    print("Nachricht: ", msg)
    print("Verschlüsselt = ",msg, "^",e," mod ",n," = ",ver)
    ent = (ver**d) % n
    print("Entschlüsselt = ",ver, "^",d," mod ",n," = ",ent)
    #signieren
    ver = (msg**d) % n
    print()
    print("Signierung1 = ",msg, "^",d," mod ",n," = ",ver)
    ent = (ver**e) % n
    print("Signierung2 = ",ver, "^",e," mod ",n," = ",ent)
```

```python
def rsa(nachricht, p, q, d=0, e=0):
    msg = nachricht
    n = p * q
    print("n = p * q = ", p, " * ",q, " = ",  n)
    e_n = euler(p,q)
    print("φ(n) = (p-1) * (q-1) = ", p-1, " * ", q-1, " = ", e_n)
    if(e == 0):
        e = find_e(e_n)
    if(d == 0):
        d = find_d(e,e_n)
    print()
    print("pub_key (n,e) = (",n,",",e,")")
    print("prv_key (n,d) = (",n,",",d,")")
    print()
    crypt(msg,e,d,n)
    print()
    print("Länge: log2 (",n,") Bit")
    print("log2 (",n,") = ", math.frexp(n)[1] - 1,"Bit")
```
